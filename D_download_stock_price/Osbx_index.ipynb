```python
import pandas as pd
import yfinance as yf
import sqlite3
```

```python
osbx = yf.download("^OSEBX", period="11y", interval="1d", )
osbx.columns = [col[0] for col in osbx.columns]
osbx = osbx.reset_index()
osbx["ticker"] = "OSEBX"
```

```python
conn = sqlite3.connect("../Nexus.db")

osbx.to_sql(
    name="osbx",
    con=conn,
    if_exists="replace",
    index=False
)

conn.close()
```
